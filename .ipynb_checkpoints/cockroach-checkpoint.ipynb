```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import csv

c_data = []
c_answer = []
prefectures = []
world_cities = []
with open("cockroach_japan.csv", 'r') as cockroach:
    reader = csv.reader(cockroach)
    tmp = [row for row in reader]
    del tmp[0:2]
    prefectures = [prefecture[0] for prefecture in tmp]
    tmp2 = [[float(s) for s in a[1:]] for a in tmp]
    c_data = np.array(tmp2).T[:-1].T
    c_answer = np.array(tmp2).T[-1]
    print(c_data.shape)
    print(c_answer.shape)
c_data_nor = (c_data - np.mean(c_data, axis=0)) / np.std(c_data, axis=0)
c_answer_nor = (c_answer - np.mean(c_answer)) / np.std(c_answer)
c_data_world = []
with open("cockroach_world.csv", 'r') as cockroach_world:
    reader = csv.reader(cockroach_world)
    tmp = [row for row in reader]
    del tmp[0:2]
    world_cities = [world_city[0] for world_city in tmp]
    tmp2 = [[float(s) for s in a[1:]] for a in tmp]
    c_data_world = np.array(tmp2)
    print(c_data.shape)
# c_data_world_nor = (c_data_world - np.mean(c_data_world, axis=0)) / np.std(c_data_world, axis=0)
for i in range(len(prefectures)):
    print("%s: %s" % (prefectures[i], c_answer_nor[i]))
print(sorted(c_answer_nor))
for i in range(len(prefectures)):
    print("%s: %s" % (prefectures[i], c_answer[i]))
```
